# Popular Data Science Questions

Our goal in this project is to use [Data Science Stack Exchange](https://datascience.stackexchange.com/) to determine what content should a data science education company create, based on interest by subject. 

## Stack Exchange

[Stack Exchange](https://stackexchange.com/sites?view=list#percentanswered) hosts sites on a multitude of fields and subjects, including mathematics, physics, languages, chess, game development and data science.

Stack Exchange has a reputation award system for its questions and answers. Each question and answer in a post is subject to upvotes and downvotes , this ensures that good contents are easily identifiable.

#### What kind of questions are welcome on this site ?

On DSSE's [help center](https://datascience.stackexchange.com/help) we can read what are the features of a good question:

* Avoid asking subjective questions
* Be on topic
* Be specific
* Make it relevant to others

#### What , other than questions , does the site's home subvided into?

On the [home page](https://datascience.stackexchange.com/) there are four sections:

* *Questions* — A list of all questions asked;
* *Tags* — A list of tags (keyword or labels that categorize the questions);

* *Users* — A list of users with the hightest reputation ordered descending;
* *Unanswered* — A list of unanswered questions;

The tags system used by Stack Exchange seems to be very useful for our aim , we can quantify how many questions are asked about each subject.

The site is heavely moderate by the community , so we can use the tagging system to derive conclusion with a good confidence.

#### What information is available in each post ?

Look ,just as an example, at [this](https://datascience.stackexchange.com/questions/6107/what-are-deconvolutional-layers) question, some of information we see:

* For both questions and answers:
    * The posts's score;
    * The posts's author;
    * The posts's body;
* For only the questions:
    * The last time the question as active;
    * How many times the question was viewed;
    * Related questions;
    * The question's tags;

## Exploring the Data

Let's start importing all the libraries that we need making sure that the *CreationDate* column will be stored as a datetime object.

````python
#Import all the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
#Read the csv file into a dataframe
questions = pd.read_csv("2019_questions.csv", parse_dates=["CreationDate"])
````

````python
#Get some information about the dataframe
questions.info()
````

We see that *FavouriteCount* column has a lot of missing values. A missing value in this column probably means that the question isn't present in any user's favourite list, so we can, without worries, replace the missing values with zero. 

The types of each column seems to be adeguate , we'll only explore deeply the *Tags* column type which is *object*.
Since the object dtype is a catch-all type, let's see what type the object in questions are.

````python
questions["Tags"].apply(lambda value: type(value)).unique()
````

We see that every value in this column is a string and choose to store the data in the *Tags* column in a list.

## Cleaning the Data

Let's begin fixing the *FavoriteCount* column.

````python
#Fix the FavoriteCount column
questions["FavoriteCount"].fillna(0, inplace=True)
questions["FavoriteCount"] = questions["FavoriteCount"].astype(int)
questions.dtypes
````

Let's now modify the *Tags* column to make it easier to work with.

````python
#Fix the Tags column to obtain a list
questions["Tags"] = questions["Tags"].str.replace("^<|>$","").str.split("><")
questions.sample(4)
````

## Most Used and Most Viewed

We now focus on determining the most popular tags. We'll do so by considering two different popularity proxies: for each tag we'll count how many times the tag was used, and how many times a question with that tag was viewed.

Let's begin counting how many times a tag was used.

````python
#Create a dict to count the tags
counting_tags = {}
for element in questions["Tags"]:
    for tag in element:
        if tag in counting_tags:
            counting_tags[tag] += 1
        else:
            counting_tags[tag] = 1            
````

For improved aesthetics, let's transform *counting_tags* in a dataframe.

````python
#Create a dataframe
counting_tags = pd.DataFrame.from_dict(counting_tags, orient="index")
counting_tags.rename(columns = {0:"Count"}, inplace=True)
counting_tags.head(10)
````

Let's now sort the dataframe to see the most used tags.

````python
#Sort the dataframe
counting_tags.sort_values(by=["Count"], inplace=True, ascending=False)
most_used = counting_tags.head(20)
most_used
````

Let's visualize the data plotting the results.
We'll show the top 20 results to have a complete vision of the situation.

````python
#Plotting the results
most_used.plot(kind="barh", figsize=(16,8))
plt.gca().invert_yaxis()
````

Some tags are very, very broad and are unlikely to be useful; e.g.: python, dataset, r. Before we investigate the tags a little deeper, let's repeat the same process for views.

````python
#Create a dict to count the views
counting_views = {}
for element in range(0,questions.shape[0]):
    for tag in questions.iloc[element][4]:
        if tag in counting_views:
            counting_views[tag] += questions.iloc[element][3]
        else:
            counting_views[tag] = questions.iloc[element][3]            
````

For improved aesthetics, let's transform counting_views in a dataframe.

````python
#Create a dataframe
counting_views = pd.DataFrame.from_dict(counting_views, orient="index")
counting_views.rename(columns = {0:"ViewCount"}, inplace=True)
counting_views.head(10)
````

Let's sort the dataframe.

````python
#Sort the dataframe
counting_views.sort_values(by=["ViewCount"], inplace=True, ascending=False)
most_viewed = counting_views.head(20)
most_viewed
````

And finally let's visualize the results.

````python
#Plotting the results
most_viewed.plot(kind="barh", figsize=(16,8))
plt.gca().invert_yaxis()
````

Let's now see them side by side.

````python
#Plotting the results side by side
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches((24, 10))
most_used.plot(kind="barh", ax=axes[0], subplots=True)
most_viewed.plot(kind="barh", ax=axes[1], subplots=True)
axes[0].invert_yaxis()
axes[1].invert_yaxis()
````

## Relation Between Tags

Now we want to determine how tags are related each other.
First of all we'll begin creating a list of all tags.

````python
#Create a list of all tags
all_tags = list(counting_tags.index)
````

Then we'll create a dataframe in which each row represent a tag and each column as well.

````python
#Create the dataframe
relations = pd.DataFrame(index=all_tags, columns=all_tags)
relations.iloc[:4,:4]
````

Now let's fill the dataframe with the relevant data for our analysis. We'll fill the dataframe with all 0 and then we'll start counting the occurences for each couple of values.

````python
#Fill the dataframe with all zeros
relations.fillna(0, inplace=True)

#Fill the dataframe with relevant values
for element in questions["Tags"]:
    for tag in element:
        for other_tags in element:
            relations.loc[tag, other_tags] += 1
````

````python
#Take the most used indices
relations_most_used = relations.loc[most_used.index, most_used.index]
relations_most_used
````

We can note that the higher values are in in the relations between machine learning, python and deep learning. However it's hard for a human to understand what's going on in this dataframe. To avoid this discomfort we'll create a heatmap which is surely more human friendly.

To have a more clear vision we'll set to zero all the relations between a tag and the tag itself.

````python
#Fix the values
for tag in most_used.index:
    relations_most_used.loc[tag, tag] = 0
````

````python
#Create the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(relations_most_used, cmap="Blues")
````

Now the situation is much more clear. The most used tags (machine learning, python, deep learning) also seem to have the strongest relationship as given by the concentration in the top left corner.

## Just a Fad ?

Before we officially made our recommendation, it would be nice to solidy our findings with additional proof. Ideally, we would like that the content we decide to create will be the most useful for as long as possible. We want to answer this question : Could interest in deep learning be slowing down ?

We'll track the interest in deep learning across time !

````python
#Read the file into a dataframe
all_q = pd.read_csv("all_questions.csv",parse_dates=["CreationDate"])

#Fix the tags column
all_q["Tags"] = all_q["Tags"].str.replace("^<|>$","").str.split("><")
````

Now we want to define the meaning of "deep learning question". First of all we should decide what tags are deep learning tags. 

The definition we'll use is: a tag is a deep learning tag if it belongs to the list ["lstm", "cnn", "scikit-learn", "tensorflow", "keras", "neural-network", "deep-learning"].

We'll now create a function that assigns 1 to deep learning questions and 0 otherwise.

````python
#Define the function
def deep_learning_q(tags):
    for tag in tags:
        if tag in ["lstm", "cnn", "scikit-learn", "tensorflow", "keras", "neural-network", "deep-learning"]:
            return 1
    return 0

#Use the function to the dataframe
all_q["DeepLearning"] = all_q["Tags"].apply(deep_learning_q)
all_q.sample(5)
````

Now we'll analyze the data across time. We'll divide each year in quarters and we'll try to understand if the interest in deep-learning is constant over time. We'll delete data in the year 2020 because we haven't data for the first quarter of 2020. 

````python
#Update the dataframe
all_q = all_q[all_q["CreationDate"].dt.year != 2020]

#Define the function
def quarter(date):
    year = str(date.year)[-2:]
    quart = str(int((date.month + 2) / 3))
    return "{}Q{}".format(year, quart)

#Apply the function
all_q["Quarter"] = all_q["CreationDate"].apply(quarter)
````

````python
#Display the data
all_q.head()
````

Now let's group by quarter and compute the ratio between the number of deep learning questions and the total number of questions.

````python
quarterly = all_q.groupby("Quarter").agg({"DeepLearning" : ["sum", "size"]})
quarterly.columns = ["DeepLearningQuestions", "TotalQuestions"]
quarterly["DeepLearningRate"] = quarterly["DeepLearningQuestions"] / quarterly["TotalQuestions"]

quarterly.reset_index(inplace=True)
quarterly.head()
````

Finally let's visualize the data.

````python
#Visualize the data
quarterly.plot(figsize=(10,10), kind="bar", x="Quarter", y="DeepLearningRate", rot=45, colormap="Pastel2")
````
